```python
def sample_by_category(df, category_col, fraction):
    return df.groupby(category_col, group_keys=False).apply(lambda x: x.sample(frac=fraction))
```

```python
sampled_df = sample_by_category(multisimlex_cn, 'PoS', 0.25)
```

```python
def get_mean_annotation(multisimlex_df):
    annotator_columns = [f'Annotator {i}' for i in range(1, 12)]
    multisimlex_df[annotator_columns] = multisimlex_df[annotator_columns].astype(float)
    avg_list = []
    
    for index, row in multisimlex_df.iterrows():
        avg_mean = round(row[annotator_columns].mean(), 2)
        avg_list.append(avg_mean)

    multisimlex_df['avg_anno'] = avg_list

    return multisimlex_df
```

```python
sampled_avg_df = get_mean_annotation(sampled_df)
```
