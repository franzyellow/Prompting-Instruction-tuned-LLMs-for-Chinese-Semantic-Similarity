```python
import pandas as pd
import numpy as np
```

```python
f1_diff = pd.read_csv('f1_diff.csv')
f3_diff = pd.read_csv('f3_diff.csv')
f5_diff = pd.read_csv('f5_diff.csv')
f6_diff = pd.read_csv('f6_diff.csv')
```

```python
type(f1_diff['pred_act_gap'])
```

```python
avg_df = pd.DataFrame({
    'word1': f1_diff['word1'],
    'word2': f1_diff['word2'],
    'pred_sl_f1': f1_diff['pred_sl'],
    'pred_sl_f3': [np.nan] * 3 + list(f3_diff['pred_sl']),
    'pred_sl_f5': f5_diff['pred_sl'],
    'pred_sl_f6': f6_diff['pred_sl'],
    'act_sl_f1': f1_diff['act_sl'],
    'act_sl_f3': [np.nan] * 3 + list(f3_diff['act_sl']),
    'act_sl_f5': f5_diff['act_sl'],
    'act_sl_f6': f6_diff['act_sl'],
})
avg_df
```

```python
pred_cols = [f'pred_sl_f{i}' for i in [1,3,5,6]]
act_cols = [f'act_sl_f{i}' for i in [1,3,5,6]]
avg_df['avg_pred_sl'] = avg_df[pred_cols].mean(axis=1).round(2)
avg_df['avg_act_sl'] = avg_df[act_cols].mean(axis=1).round(2)

avg_df['avg_pred_act_gap'] = avg_df['avg_pred_sl'] - avg_df['avg_act_sl']
avg_df_cleaned = avg_df[['word1', 'word2','avg_pred_sl', 'avg_act_sl', 'avg_pred_act_gap']]
avg_df_cleaned_sorted = avg_df_cleaned.sort_values(by = 'avg_pred_act_gap', ascending = False)
avg_df_cleaned_sorted
```

```python
avg_df_cleaned_sorted.to_csv('average_diff.csv')
```
