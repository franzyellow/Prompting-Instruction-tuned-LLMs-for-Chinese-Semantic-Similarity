```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img1 = Image.open("f1sim.png")
img2 = Image.open("f3sim.png")
img3 = Image.open("f5sim.png")
img4 = Image.open("f6sim.png")
img5 = Image.open("f9sim.png")

#计算每张图片的宽度和高度
img_width, img_height = img1.size  # 假设所有图片大小一致

# 计算新画布的尺寸
columns = 3  # 下面一行 3 张
rows = 2  # 上面一行 2 张
canvas_width = img_width * columns
canvas_height = img_height * rows

# 创建一个新画布（白色背景）
merged_img = Image.new("RGB", (canvas_width, canvas_height), (255, 255, 255))

# 逐个粘贴图片到新画布上
# 第一行（2 张）
merged_img.paste(img1, (0, 0))  # 左上角
merged_img.paste(img2, (img_width, 0))  # 右上角

# 第二行（3 张）
merged_img.paste(img3, (0, img_height))  # 左下角
merged_img.paste(img4, (img_width, img_height))  # 中间
merged_img.paste(img5, (img_width * 2, img_height))  # 右下角

# 保存最终拼接后的大图
merged_img.save("big_boxplot.png")

# 显示图片（可选）
merged_img.show()
image_paths = ["f1sim.png", "f3sim.png", "f5sim.png", "f6sim.png", "f9sim.png"]
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# 确保 5 张图片存在
image_paths = ["f1sim.png", "f3sim.png", "f5sim.png", "f6sim.png", "f9sim.png"]
missing_files = [img for img in image_paths if not os.path.exists(img)]

if missing_files:
    print("❌ 缺失图片:", missing_files)
else:
    print("✅ 所有图片都存在")

# 创建 2×3 网格
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

# 移除 **最后一个 subplot**（因为只有 5 张图，不是 6 张）
fig.delaxes(axes[1, 2])

# 依次填充 5 张图片
for i, img_path in enumerate(image_paths):
    img = mpimg.imread(img_path)
    axes.flat[i].imshow(img)
    axes.flat[i].axis("off")  # 移除坐标轴

# 调整布局
plt.tight_layout()

# 保存最终合成的图片
plt.savefig("big_boxplot.png", dpi=300, bbox_inches="tight")

# 显示图片
plt.show()
```
